# Top-Down Design

Remember early in the semester when I said algorithmic thinking is at the core of computer science? A computer scientist has three types of thing in their skills kit:

1. A set of tools for solving problems (like lego pieces). These tools are **data structures** (like lists, dictionaries and trees!) and **algorithms** (like iteration, sort and search!).
2. One or more ways to write those tools down (a natural language like English, pseudocode, python, Java....)
3. A structured process for defining a big problem and breaking it down into smaller problems that can be solved using the tools they know. This is typically done using **top-down design**.

**Top-down design** is "a method of breaking a problem down into smaller, less complex pieces from the initial overall problem" (from https://computer-based-problem-solving.readthedocs.io/en/latest/problem-solving/top-down-design.html).
